import pandas as pd

##Displaying Data:

**.head(n):** Displays the first n rows of the DataFrame.

**.tail(n)**: Displays the last n rows of the DataFrame.

**.sample(n):** Randomly selects n rows from the DataFrame.

**.info():**  Provides information about the DataFrame's structure

**.describe():** Generates summary statistics of the data.

##Filtering Data:

**.loc[]:** Allows for label-based indexing and slicing.

**.iloc[]:** Allows for integer-based indexing and slicing.

**.query():** Allows you to query the DataFrame using a boolean expression.
Data Transformation:

**.drop():** Removes specified columns or rows.

**.sort_values():** Sorts the DataFrame by one or more columns.

**.groupby():** Groups the data based on one or more columns.

**.pivot_table():** Creates a pivot table for data summarization.

**.apply():** Applies a function to each element or row/column in the DataFrame.
Data Cleaning:

**.fillna():** Fills missing values with specified data.

**.dropna():** Removes rows or columns with missing values.

**.replace():** Replaces values in the DataFrame.

##Data Aggregation:

**.sum():** Calculates the sum of values in a column.
**.mean():** Computes the mean of values in a column.
**.count():** Counts the number of non-null values in a column.
Merging and Joining DataFrames:

**.merge():** Combines two DataFrames based on a common column.
**.concat():** Concatenates DataFrames along rows or columns.

##Plotting and Visualization:

**.plot():** Generates various types of plots and charts.
Saving Data:

**.to_csv():** Exports the DataFrame to a CSV file.

**.to_excel():** Exports the DataFrame to an Excel file.

##Pandas Series

```python
#pandas series like a coloumn in a table
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)

#Labels
#First value has index 0, second value has index 1 etc.
print(myvar[1])
```

```python
#Create Labels
# here we can change coloumn name
a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
print(myvar["z"])
```

###Key/Value Objects as Series

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

##DataFrames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

Series is like a column, a DataFrame is the whole table.

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data, index = ["x", "y", "z"])

print(myvar)
```

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

##Locate Row

As you can see from the result above, the DataFrame is like a table with rows and columns.

Pandas use the **loc** attribute to return one or more specified row(s)

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)

```

```python
#refer to the row index:
print(df.loc[0])
```

```python
#eturn row 0 and 1:
print(df.loc[[0,1]])
```

```python
#Add a list of names to give each row a name:
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df)
```

```python
#refer to the named index:
print(df.loc["day2"])
```

```python
# Here we are import dataset
df = pd.read_csv('/content/drive/MyDrive/dataset_encome/archive/dataset.csv')

print(df.to_string())
```

```python
#Calculate the MEAN, and replace any empty values with it:
x = df["cpi_country"].mean()

df["cpi_country"].fillna(x, inplace = True)
print(x)
```

```python
#MEDIAN, and replace any empty values with it:
x = df["cpi_country"].median()

df["cpi_country"].fillna(x, inplace = True)
print(x)
```

```python
#MODE
x = df["cpi_country"].mode()[0]

df["cpi_country"].fillna(x, inplace = True)
print(x)
```
